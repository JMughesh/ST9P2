```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
```

```python
# Load dataset and reduce to 2 features for visualization
data = load_iris()
X = data.data[:, :2]  # Only first 2 features (sepal length & width)
y = data.target
```

```python
# Only keep 2 classes (binary classification)
X = X[y != 2]
y = y[y != 2]
```

```python
# Train linear SVM
model = SVC(kernel='linear')
model.fit(X, y)
```

```python
# Plot
plt.figure(figsize=(8,6))
```

```python
# Plot all data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k', s=60)
```

```python
# Plot support vectors
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=150, facecolors='none', edgecolors='black', label='Support Vectors')
```

```python
# Create grid for decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                     np.linspace(*ylim, num=200))
xy = np.vstack([xx.ravel(), yy.ravel()]).T
Z = model.decision_function(xy).reshape(xx.shape)
```

```python
# Plot decision boundary and margins
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1],
            linestyles=['--', '-', '--'])
plt.title("SVM with Linear Kernel and Margin Visualization")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
